# 테스트 영상 보기

```python
import glob
from IPython.display import HTML
from base64 import b64encode
import os

inpFile = f"{globalVar['outPath']}/result_20240710/*.mp4"
fileList = sorted(glob.glob(inpFile, recursive=True))
fileList

if fileList is None or len(fileList) < 1:
    print(f'inpFile : {inpFile} / 입력 자료를 확인해주세요')

for fileInfo in fileList:
    fileName = os.path.basename(fileInfo)
    # fileNameNoExt = os.path.basename(fileInfo).split('.')[0]

    saveFileInfo = fileInfo
    os.makedirs(os.path.dirname(saveFileInfo), exist_ok=True)
    print(f"[CHECK] saveFileInfo : {saveFileInfo}")

    comFileInfo = f"{globalVar['outPath']}/com_20240710/{fileName}"
    os.makedirs(os.path.dirname(comFileInfo), exist_ok=True)
    print(f"[CHECK] comFileInfo : {comFileInfo}")

    cmd = f"ffmpeg -y -i {saveFileInfo} -vcodec libx264 {comFileInfo}"
    print(f"[CHECK] cmd : {cmd}")
    ! {cmd}

    # Show video
    mp4 = open(comFileInfo,'rb').read()
    urlVid = "data:video/mp4;base64," + b64encode(mp4).decode()
    HTML("""
    <video width=1000 controls autoplay>
        <source src="%s" type="video/mp4">
    </video>
    """ % urlVid)
```
