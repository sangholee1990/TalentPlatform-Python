# 開始import資料

```python
import numpy as np 
import pandas as pd
import os

df = pd.read_csv("london_merged.csv")
df.head()
```

# Data Pre-processing

```python
#Check Missing Vaule & Column Range // can skip running this block of code when open next time 
```

```python
import pandas_profiling as pp
profile = pp.ProfileReport(df)
profile
```

```python
# Seperate timestamp into "month" "week" "day" "hour"  // year = 2015 so no need to add 
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.set_index("timestamp")
df.head()
```

```python
# Seperate timestamp into "month" "week" "day" "hour" 
```

```python
df["hour"] = df.index.hour
df["day_of_month"] = df.index.day
df["day_of_week"]  = df.index.dayofweek
df["month"] = df.index.month
df.head()
```

```python
# we can see "month" "week" "day" "hour" in seperate columns above
# import library
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Overall Trend
```

```python
sns.set_palette("pastel")
plt.figure(figsize=(15, 7))
ax = sns.lineplot(x=df.index, y=df.cnt,data=df)
#ax = sns.relplot(x=df.index, y=df.cnt,data=df, hue="month", kind="line")
ax.set_title("Amount of bike shares vs date", fontsize=25)
ax.set_xlabel("Date", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.show()
```

```python
# Month Trend
```

```python
df_by_month = df.resample(rule = "M").sum()

sns.set_palette("dark")
plt.figure(figsize=(16,6))
ax = sns.lineplot(data=df_by_month,x=df_by_month.index,y=df_by_month.cnt)
ax.set_title("Amount of bike shares per month", fontsize=25)
ax.set_xlabel("Month", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.show()
```

```python
# Season Trend
```

```python
plt.figure(figsize=(16,6))
ax = sns.pointplot(data=df,hue=df.season,y=df.cnt,x=df.month)
ax.set_title("Amount of bike shares per season", fontsize=25)
ax.set_xlabel("Month", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.show()
```

```python
# Weekly Trend
```

```python
plt.figure(figsize=(16, 6))
ax = sns.pointplot(x='day_of_week', y='cnt',data=df)
ax.set_title("Amount of bike shares in a week", fontsize=25)
ax.set_xlabel("Day of the week", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.show()
```

```python
# Daily Trend
```

```python
df_by_week = df.resample("D").sum()
plt.figure(figsize=(16,6))
ax = sns.lineplot(data=df_by_week,x=df_by_week.index,y=df_by_week.cnt)
ax.set_title("Amount of bike shares per day", fontsize=25)
ax.set_xlabel("Time", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.show()
```

```python
# Hour Trend in Day // consider holiday
```

```python
plt.figure(figsize=(16, 6))
ax = sns.pointplot(x='hour', y='cnt',hue='is_holiday',data=df)
ax.set_title("Amount of bike shares per hour in a day", fontsize=25)
ax.set_xlabel("Hour of they day", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.show()
```

```python
# Hour Trend // Consider weekend
```

```python
plt.figure(figsize=(16, 6))
#ax = sns.pointplot(x='hour', y='cnt',data=df[df["is_weekend"]==1]) # just plot weekend day
ax = sns.pointplot(x='hour', y='cnt',data=df,hue='is_weekend')
ax.set_title("Amount of bike shares per hour on a weekend day", fontsize=25)
ax.set_xlabel("Hour of they day", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.show()
```

```python
# Temperature Trend
```

```python
plt.figure(figsize=(20,10))

ax = sns.lineplot(x='t1', y='cnt',data=df, ci=None) # ci = None 取消顯示可信區間
ax.set_title("Amount of bike shares vs real temperature", fontsize=25)
ax.set_xlabel("Real temperature (°C)", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.locator_params(axis='x', nbins=10)
```

```python
plt.figure(figsize=(20,10))

ax = sns.lineplot(x='t2', y='cnt',data=df, ci=None) # ci = None 取消顯示可信區間
ax.set_title("Amount of bike shares vs Feel temperature", fontsize=25)
ax.set_xlabel("Feel temperature (°C)", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.locator_params(axis='x', nbins=10)
```

```python
# humidity relation
```

```python
corr_t1_humid = df["t1"].corr(df["hum"], method = "pearson")
print("corr: ",corr_t1_humid)
```

```python
plt.figure(figsize=(20,10))

ax = sns.pointplot(x='hum', y='cnt',data=df,color="#2a9d8f")
ax.set_title("Amount of bike shares vs humidity", fontsize=25)
ax.set_xlabel("Humidity (%)", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.locator_params(axis='x', nbins=10)
plt.show()
```

```python
# wind speed
```

```python
plt.figure(figsize=(20,10))

ax = sns.pointplot(x='wind_speed', y='cnt',data=df,color = '#f72585')
ax.set_title("Amount of bike shares vs windspeed", fontsize=25)
ax.set_xlabel("Windspeed (km/h)", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.locator_params(axis='x', nbins=10)
plt.show()
```

```python
corr_t1_wind_speed = df["t1"].corr(df["wind_speed"], method = "pearson")
print("corr: ",corr_t1_wind_speed)
```

```python
# weather condition relation
```

```python
plt.figure(figsize=(20,6))
plate = sns.color_palette("flare", as_cmap=True)
ax = sns.histplot(data=df,y=df.cnt,x=df.weather_code,cbar=False,hue = 'weather_code',palette = plate)
ax.set_title("Amount of bike shares vs the weather", fontsize=25)
ax.set_xlabel("Weather code", fontsize=20)
ax.set_ylabel('Amount of bike shares', fontsize=20)
plt.show()
```
