```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
from xgboost.sklearn import XGBRegressor
from xgboost import plot_importance
import time
from sklearn.metrics import mean_absolute_error,r2_score, mean_squared_log_error,mean_squared_error, make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import RobustScaler
```

```python
data = pd.read_csv("london_merged.csv")
X = data.drop(['timestamp','cnt','t2'],axis=1)
y = data['cnt']

def df_split(df,train_percent):
    split_index = int(train_percent * len(df))
    train = df.iloc[:split_index]
    test = df.iloc[split_index:]
    return train,test

X_train,X_test = df_split(X,0.9)
y_train,y_test = df_split(y,0.9)

print(y_test)

# scaler = RobustScaler()

# X = X_train[:,0:2]
# X2 = y_train[:,0:2]
# X_train_new= scaler.fit_transform(X)
# y_train_new = scaler.fit_transform(X2)

# X_test['cnt'] = scaler.fit_transform(X_test[:,'cnt'])
# y_test['cnt'] = scaler.fit_transform(y_test[:,'cnt'])

```

```python
# tune
```

```python
cv_params = {'learning_rate': [0.043, 0.044, 0.045, 0.046, 0.047]}
other_params = {'learning_rate': 0.045, 'n_estimators': 350, 'max_depth': 3, 'min_child_weight': 6, 'seed': 0,
                'subsample': 0.6, 'colsample_bytree': 0.6, 'gamma': 0.08, 'reg_alpha': 3, 'reg_lambda': 0.1}

model = XGBRegressor(**other_params)
optimized_GBM = GridSearchCV(estimator=model, param_grid=cv_params, scoring='neg_root_mean_squared_error', cv=5, verbose=1,n_jobs=4)
optimized_GBM.fit(X_train, y_train)
evalute_result = optimized_GBM.cv_results_
print('每輪迭代執行結果:{0}'.format(evalute_result))
print('引數的最佳取值：{0}'.format(optimized_GBM.best_params_))
print('最佳模型得分:{0}'.format(optimized_GBM.best_score_))
```

```python
# final
```

```python
xgbmodel_fine = XGBRegressor(learning_rate= 0.046, n_estimators= 350, max_depth= 3, min_child_weight= 6, seed= 0,
                subsample= 0.6, colsample_bytree= 0.6, gamma= 0.08, reg_alpha= 3, reg_lambda= 0.1)
t1 = time.time()
xgbmodel_fine.fit(X_train,y_train)
t2 = time.time()
print('Train time: ',t2-t1)

t1 = time.time()
preds = xgbmodel_fine.predict(X_test)
t2 = time.time()
print('Predict time: ',t2-t1)
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))

r2 =  r2_score(y_test, preds)
print('R2: ',r2)

plt.figure(figsize=(16,6))
plt.plot(y_test.values,marker=".",label="actual",color="#eee8aa")
plt.plot(preds,marker=".",label="prediction",color="#00bfff")

plt.title('Bike share prediction per hour', fontsize=22)
plt.xlabel(xlabel='Date', fontsize=14)
plt.ylabel(ylabel='Bikes shares', fontsize=14)
plt.legend(['Ground truth', 'Predicted'])

plt.show()
```
