# OrderData 로드

```python
import pandas as pd
df = pd.read_excel('processed_orderdata.xlsx')
```

```python
statesAbbrev = pd.read_excel('statesAbbrev.xlsx')
statesAbbrev.dtypes
```

```python
df.head()
df.dtypes
```

```python

df =pd.merge(df, statesAbbrev, how='inner', on='Shipment To State')
# df.join(statesAbbrev, on='Shipment To State', how='inner')
```

```python
df
```

```python
# 0-4 세 영유아 2019년 데이터
childPopData = pd.read_excel('child_data.xlsx')
only_2019 = childPopData['TimeFrame'] == 2019
childPopData = childPopData[only_2019]
childPopData['State Full Name'] = childPopData['State Full Name'].str.upper()
childPopData
```

# 구매데이터 + 0-4세 영유아 데이터 MERGE

```python
orderdata_with_child_popdata = pd.merge(df, childPopData, on='State Full Name', how='inner')
```

```python
orderdata_with_child_popdata.to_excel('OrderData + ChildData.xlsx', index=False)
```
