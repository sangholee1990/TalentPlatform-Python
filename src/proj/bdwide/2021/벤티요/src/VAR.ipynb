# 데이터셋 로드

```python
# load dataset
df = pd.read_csv('final_merged_cali.csv', header=0, index_col=0)
df= df.fillna(df.mean())
df['date_time'] = pd.to_datetime(df.date_time , format = '%Y-%m-%d')
data = df.drop(['date_time'], axis=1)
data.index = df.date_time
data.drop(["Population"], axis=1, inplace=True)
data.astype(float)
# values = data.values
```

```python
data.tail()
```

# 훈련 및 검증 데이터 셋 분리 및 VAR 모델 생성

```python
#creating the train and validation set
train = data[:int(0.8*(len(data)))]
valid = data[int(0.8*(len(data))):]


#fit the model
from statsmodels.tsa.vector_ar.var_model import VAR

model = VAR(endog=train)
model_fit = model.fit()

# make prediction on validation
prediction = model_fit.forecast(model_fit.y, steps=len(valid))
```

```python
#converting predictions to dataframe
from math import sqrt
from sklearn.metrics import mean_squared_error
cols = data.columns
print(cols)
pred = pd.DataFrame(index=range(0,len(prediction)),columns=[cols])

for j in range(0,6):
    for i in range(0, len(prediction)):
       pred.iloc[i][j] = prediction[i][j]

pred.head()
```

```python
#check rmse
import numpy as np

for i in cols:
    print(pred[i], valid[i])
    # print('rmse value for', i, 'is : ',mean_squared_error(pred[i], valid[i]))
```
