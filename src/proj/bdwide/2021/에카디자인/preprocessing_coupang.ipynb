* KSA is a korean sentiment analyzer for general korean texts. KSA는 일반적인 한국어 감성분석 용도로 사용합니다. 
* 형태소 분석기는 서울대학교 IDS 연구실에서 만든 꼬꼬마를 사용한다. 감성사전 또한 동 연구소의 것을 사용한다. (참고: http://kkma.snu.ac.kr/)

```python
from ekonlpy.sentiment import KSA
ksa = KSA()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction import DictVectorizer
from scipy.sparse import hstack
from sklearn.linear_model import Ridge

df = pd.read_csv("total.csv", encoding="cp949")
df
```

## 리뷰내용 감정분석

```python
# postive : positive로 예상되는 단어개수
# negative : negative로 예상되는 단어개수
sentiments = []
for row in df['리뷰내용']:
   tokens = ksa.tokenize(row)
   score = ksa.get_score(tokens)
   sentiments.append(score['Polarity']) 

df['쿠팡리뷰감정지수'] = sentiments
```

## 리뷰내용에서 키워드 추출

```python
# -*- coding: utf-8 -*-
from krwordrank.word import KRWordRank


def extractKeywords(texts, num_keywords=20):
    wordrank_extractor = KRWordRank(min_count=3, max_length=10)
    keywords, rank, graph = wordrank_extractor.extract(texts, num_keywords=num_keywords)
    return keywords
```

```python
# 키워드 추출
keywords_list = []
for row in df['리뷰내용']:
    keywords = extractKeywords([row])
    keywords = ' '.join(keywords.keys())
    keywords_list.append(keywords)

df['키워드'] = keywords_list

```
