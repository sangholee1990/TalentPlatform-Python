## 라이브러리 IMPORT

```python
import pandas as pd
from ekonlpy.sentiment import KSA
ksa = KSA()
```

## 네이버 리뷰 데이터 불러오기

```python
df = pd.read_csv("naver_jisikin_15_08_2021 215723.csv")
df.head()
```

## question + answers 컬럼 생성 -> 후에 텍스트 감정 분석

```python
df['question+answers'] = df['question'] + df['answers']
df.head()
# ksa.tokenize()
```

## KSA 클래스 활용 토크나이징 및 감정분석.

```python
sentiments = []
for row in df["question+answers"]:
    tokens = ksa.tokenize(row)
    score = ksa.get_score(tokens)
    sentiments.append(score['Polarity'])

df['sentiments'] = sentiments
```

```python
df.head()
```

## 리뷰데이터 검색어 기준 group_by 후 csv 저장

```python
df.groupby(["search_keyword"]).mean().to_csv("sentiments_by_keywords.csv")
```
