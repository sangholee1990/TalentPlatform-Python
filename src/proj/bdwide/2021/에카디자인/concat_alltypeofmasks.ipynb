```python
import pandas as pd
import os


maskdata = [f for f in os.listdir() if f.endswith('.csv') and f.startswith(('filterchangeable','kf94','sports','fashion','dental'))]
total_df = pd.DataFrame(data=[['','','','','','']], columns=['상품명', '상품가격', '상품종류', '리뷰제목', '리뷰내용', '별점'])
print(maskdata)
for d in maskdata:
    df = pd.read_csv(d)
    total_df = pd.concat([total_df, df])
total_df
```

```python
# csv save
total_df.to_csv("total.csv")
```
