```python
url_origin = 'https://search.shopping.naver.com/search/category?catId=50001888&frm=NVSHCAT&iq=%EB%96%A1%EB%B3%B6%EC%9D%B4&origQuery&pagingIndex={}&pagingSize=40&productSet=total&query&sort=rel&timestamp=&viewType=list'
href_list = []
brands = []
regis = []
```

```python
for i in range(5, 20):
    url = url_origin.format(i)
    driver = webdriver.Chrome(executable_path='chromedriver')
    driver.get(url)
    for q in range(1,3):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)

    html_source = driver.page_source
    soup = BeautifulSoup(html_source, 'html.parser')
    ee = soup.find('ul', {'class': 'list_basis'})
    ss = ee.find_all('div', {'class' : 'basicList_inner__eY_mq'})
    
    
    p_link = ''
    brand = ''
    for q in ss:        
        qq = str(q)
        if 'basicList_mall_list__vIiQw' in qq:            
            print('not smartstore')            
        elif 'basicList_mall_window__1t6r3'in qq:
            print('not smartstore')
        else:
            p_link = q.find('a', {'class': 'basicList_link__1MaTN'}).get('href')
        
        
        etc = q.find_all('span', {'class': 'basicList_etc__2uAYO'})
        
        
        regi = ''
        for r in etc:
            registri = r.get_text()
            if '등록일' in registri:
                regi = registri
    

        try:
            brand = q.find('a', {'class': 'basicList_mall__sbVax'}).get_text()
        except:
            print('')
            
    
        if brand == '마켓컬리':
            print('')
        elif brand == '배민상회':
            print('')
        elif brand == '':
            print('')
        elif p_link != '':
            href_list.append(p_link)
            brands.append(brand)
            regis.append(regi)
            
        p_link = ''
        brand = ''
            
            
    time.sleep(2)    
    driver.close()
    
print(len(href_list))
```

```python
def datas(soup):
    
    res = []
    name = soup.find('div', {'class': 'CxNYUPvHfB'})
    name_f = name.get_text()

    price = soup.find('strong', {'class': 'aICRqgP9zw'}).get_text()
    price_f = int(re.sub(r'[^0-9]', '', price))
    
    
    img_url = ''
    
    try:
        img_url = soup.find('img', {'class': '_2P2SMyOjl6'}).get('src')
    except:
        print('이미지 못찾음')

    review = 0   
    grade = 0
    try:
        extra = soup.find_all('strong', {'class': '_2pgHN-ntx6'})
        review = extra[0].get_text()
        review = int(re.sub(r'[^0-9]', '', review))
        
        grade = extra[1].get_text()
        grade = float(grade.split('/')[0])
    except:
        print('리뷰와 등급이 없음')
       
    
    res.append(name_f)
    res.append(price_f)
    res.append(review)
    res.append(grade)
    res.append(img_url)
    return(res)
```

```python
for i in range(0, len(href_list)):
    
    url = href_list[i]
    brand = brands[i]
    regi = regis[i]
    
    driver = webdriver.Chrome(executable_path='chromedriver')
    driver.get(url)
    check1 = driver.current_url
    
    
        
    if 'https://smartstore.naver.com/' in check1:
        
        print(i)
        product_num = check1.rsplit('/', 1)[1]
        
        for q in range(1,3):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(1)
        html_source = driver.page_source
        soup = BeautifulSoup(html_source, 'html.parser')        
        res = datas(soup)
        
        if '?' in  product_num:
            product_num = product_num.split('?', 1)[0]
            
        
        
        product_num = int(product_num)   
        
        
        
        res.append(product_num)
        res.append(brand)        
        res.append(regi)
        res.append(check1)
        
        print(res)
        
        df.loc[len(df)] = res
        
    driver.close()
    time.sleep(2)
```
