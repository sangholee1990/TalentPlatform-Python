```python
for i in INDATA:
    target_df = pd.DataFrame()
    print(i)
    f = open(i, 'rt', encoding='UTF8')
    fn = i.replace("IN","OUT").replace(".json",".xlsx")

    print(fn)
    data = json.load(f)
    size = len(data["segments"])
    
    Index = 0
    for j in range(0,size,1):
        Index = Index + 1
        Modification1 = "인식결과"
        Modification2 = "수정결과"
        Speaker1 = data["segments"][j]["speaker"]["label"]
        Speaker2 = data["segments"][j]["speaker"]["name"]
        StartTime = data["segments"][j]["start"]
        EndTime = data["segments"][j]["end"]
        Script1 = data["segments"][j]["text"]
        Script2 = data["segments"][j]["textEdited"]
        
        part1 = {
                "Index" : Index,
                "Modification" : Modification1,
                "Speaker" : Speaker1,
                "Start Time" : StartTime,
                "End Time" : EndTime,
                "Script" : Script1
                }
            
        part2 = {
                "Index" : Index,
                "Modification" : Modification2,
                "Speaker" : Speaker2,
                "Start Time" : StartTime,
                "End Time" : EndTime,
                "Script" : Script2
                }
            
        target_df = target_df.append(pd.DataFrame.from_dict([part1]))
        target_df = target_df.append(pd.DataFrame.from_dict([part2]))

    target_df.to_excel(fn,index=False)
    
```
